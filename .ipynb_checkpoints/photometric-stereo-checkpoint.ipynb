### Photometric Stereo Implementation
  
In this section, I will implement the photometric stereo technique. The program will have two parts:
<ul>
<li>Read in the images and corresponding light source directions, and estimate the surface normals and albedo map. </li>
    <li>Reconstruct the depth map from the surface with the implementation of the Horn integration technique is given below in `horn_integrate` function.</li></ul>    
<br/>

```python
import numpy as np
from scipy.signal import convolve
from numpy import linalg

def horn_integrate(gx, gy, mask, niter):
    """
    horn_integrate recovers the function g from its partial 
    derivatives gx and gy. 
    mask is a binary image which tells which pixels are 
    involved in integration. 
    niter is the number of iterations. 
    typically 100,000 or 200,000, 
    although the trend can be seen even after 1000 iterations.
    """
    g = np.ones(np.shape(gx))
    
    gx = np.multiply(gx, mask)
    gy = np.multiply(gy, mask)
    
    A = np.array([[0,1,0],[0,0,0],[0,0,0]]) #y-1
    B = np.array([[0,0,0],[1,0,0],[0,0,0]]) #x-1
    C = np.array([[0,0,0],[0,0,1],[0,0,0]]) #x+1
    D = np.array([[0,0,0],[0,0,0],[0,1,0]]) #y+1
    
    d_mask = A + B + C + D
    
    den = np.multiply(convolve(mask,d_mask,mode="same"),mask)
    den[den == 0] = 1
    rden = 1.0 / den
    mask2 = np.multiply(rden, mask)
    
    m_a = convolve(mask, A, mode="same")
    m_b = convolve(mask, B, mode="same")
    m_c = convolve(mask, C, mode="same")
    m_d = convolve(mask, D, mode="same")
    
    term_right = np.multiply(m_c, gx) + np.multiply(m_d, gy)
    t_a = -1.0 * convolve(gx, B, mode="same")
    t_b = -1.0 * convolve(gy, A, mode="same")
    term_right = term_right + t_a + t_b
    term_right = np.multiply(mask2, term_right)
    
    for k in range(niter):
        g = np.multiply(mask2, convolve(g, d_mask, mode="same")) + term_right
    
    return g
```

### Partial Derivatives

For computing partial derivatives, I will assume the following convention:

\begin{eqnarray}
\frac{\partial z}{\partial x} = \frac{\hat{n}_1}{\hat{n}_3} \\ 
\frac{\partial z}{\partial y} = \frac{\hat{n}_2}{\hat{n}_3}
\end{eqnarray}

```python
def photometric_stereo(images, lights, mask, horn_niter=25000):
    
    """mask is an optional parameter which can be used 
    e.g. to ignore the background when integrating the normals.
    It should be created by converting the images to grayscale and thresholding
    (only using locations for which the pixel value is above some threshold).
    
    The choice of threshold being around 0.05 tends to work well.
    """

    # note:
    # images : (n_ims, h, w)
    # lights : (n_ims, 3)
    # mask   : (h, w)
    
    albedo = np.ones(images[0].shape)
    normals = np.dstack((np.zeros(images[0].shape),
                         np.zeros(images[0].shape),
                         np.ones(images[0].shape)))

    H_horn = np.ones(images[0].shape)
    
    partial_x = np.zeros(images[0].shape)
    partial_y = np.zeros(images[0].shape)
        
    # 1 - NEED TO ESTIMATE NORMALS AND ALBEDO MAP
    # 2 - Reconstruct the depth map from the surface with the implementation of the Horn integration technique 
    # given above in `horn_integrate` function. Note that you will typically want to run the `horn_integrate` function 
    # with 10000 - 100000 iterations, meaning it will take a while.
    
    # b(x, y) = S^-1 e(x, y)
    
    n_row = 0
    n_col = 0
    row_len = len(images[0])
    col_len = len(images[0][0])
    b_mat = np.zeros(shape=(row_len, col_len), dtype=object )
    
    # SOLVE LINEAR SYSTEM FOR UNKNOWNS - B(x, y)
    # S - light sources (lights)
    # e - images (images[0] -> images[3])
    # b - unknown
    for r in range(row_len):
        for c in range(col_len):
            # define variables
            e_mat = []
            for n in range(len(images)):
                e_mat.append(np.array([images[n][r][c]]))
            e_mat = np.array(e_mat)
            
            if (len(images) < 4):
                S_inv = np.linalg.inv(lights)
            else:
                S_inv = np.linalg.pinv(lights)
            
            # solve for b(x, y)
            b_mat[r][c] = np.matmul(S_inv, e_mat)
            
            # solve for surface normal
            normals[r][c] = b_mat[r][c].reshape(3,) / np.linalg.norm(b_mat[r][c])
            
            # solve for albedo
            albedo[r][c] = np.linalg.norm(b_mat[r][c])
            
            # solve for partial derivatives at each pixel
            partial_x[r][c] = normals[r][c][0]/normals[r][c][2]
            partial_y[r][c] = normals[r][c][1]/normals[r][c][2]

    # solve for H_horn
    H_horn = horn_integrate(partial_x, partial_y, mask, horn_niter)
    
    return albedo, normals, H_horn
```

### Calculating Surface normals

Calculate the surface normals and display them as a quiver plot. The surface normals are given by 
\begin{eqnarray}
[-\frac{\delta f}{\delta x}, -\frac{\delta f}{\delta y}, 1].
\end{eqnarray}

```python
# Compute and plot the surface normals
# (make sure that the normal vector is normalized to unit length)
import scipy

normals = np.stack((np.zeros(heightmap.shape),np.zeros(heightmap.shape),np.ones(heightmap.shape)), axis=-1)

# This part of the code should compute the normals i.e fill in the above normals variable correctly. Then just 
# execute the next cell to plot the surface normals

row_len = len(albedo)
col_len = len(albedo[0])

# for 2d convolution use square kernels
k_x = 1/2 * np.array([[0, 0, 0], [1, 0, -1], [0, 0, 0]])
k_y = 1/2 * np.array([[0, 1, 0], [0, 0, -0], [0, -1, 0]])
conv_x = np.zeros(shape=(row_len, col_len), dtype=object)
conv_y = np.zeros(shape=(row_len, col_len), dtype=object)

# add a border padding to albedo
heightmap_pad = np.vstack((np.zeros((1, col_len)), heightmap, np.zeros((1, col_len))))
heightmap_pad = np.concatenate((np.zeros((row_len+2, 1)), heightmap_pad, np.zeros((row_len+2, 1))), axis=1)

# loop through image to separate into 3x3 matrices
# then calculate convolution on them and store in respective matrix
for r in range(row_len):
    for c in range(col_len):
        # build out matrix
        curr_mat = np.matrix([[heightmap_pad[r][c],   heightmap_pad[r][c+1],   heightmap_pad[r][c+2]],
                              [heightmap_pad[r+1][c], heightmap_pad[r+1][c+1], heightmap_pad[r+1][c+2]],
                              [heightmap_pad[r+2][c], heightmap_pad[r+2][c+1], heightmap_pad[r+2][c+2]]])
        
        # calculate convolutions
        conv_x[r][c] = scipy.signal.convolve2d(curr_mat, k_x, mode='valid')[0][0]
        conv_y[r][c] = scipy.signal.convolve2d(curr_mat, k_y, mode='valid')[0][0]
        
        # fill in normals matrix
        norm = np.array([-conv_x[r][c], -conv_y[r][c], 1])
        normals[r,c] = norm / np.linalg.norm(norm) #normalize the normal

# iterate over a 3x3 block on the image and apply convolution with respective kernels
```

```python
# This code is used to visualise the surface normals plot. 

stride = 5
X, Y = np.meshgrid(np.arange(0,np.shape(normals)[0], stride),
                          np.arange(0,np.shape(normals)[1], stride))

Z = heightmap[::stride,::stride].T
NX = -normals[...,0][::stride,::-stride].T
NY = -normals[...,1][::-stride,::stride].T
NZ = normals[...,2][::stride,::stride].T
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection = '3d')
plt.quiver(Y,X,Z,NX,NY,NZ, length=10)
plt.show()
```

### Render images

For each of the two albedos, render three images. One for each of the two light sources, and one for both light-sources combined. 
Recall that the general image formation equation is given by
\begin{eqnarray}
I = a(x,y) \hat{\boldsymbol{\mathrm{n}}}(x,y)^\top \hat{\boldsymbol{\mathrm{s}}}(x,y) s_0
\end{eqnarray}
where $a(x,y)$ is the albedo for pixel $(x, y)$, $\hat{n}(x,y)$ is the corresponding surface normal, $\hat{s}(x,y)$ the light source direction, $s_0$ the light source intensity. Let the light source intensity be $1$ and make the `distant light source assumption'.

```python
# Render Images

def lambertian(normals, light, albedo, intensity, mask):
    
    row_len = len(albedo)
    col_len = len(albedo[0])
    img_mat_1 = np.zeros(shape=(row_len, col_len), dtype=float )
    img_mat_2 = np.zeros(shape=(row_len, col_len), dtype=float )
    img_mat_3 = np.zeros(shape=(row_len, col_len), dtype=float )
    
    # SOLVE LINEAR SYSTEM FOR UNKNOWNS - Img
    # S - light sources (lights)
    # a - albedo at (x, y)
    # n - normals at (x, y)
    # s0 - light intensity (in this case constant 1)
    # I - unknown
    for r in range(row_len):
        for c in range(col_len):
            # solve for variables
            a_n = albedo[r][c] * normals[r][c]
            a_n = a_n.T
            s_mat1 = intensity * light[0]
            s_mat2 = intensity * light[1]
            s_mat3 = intensity * light[2]
            
            # solve for I(x, y)
            # clip to ensure values between [0, 1]
            img_mat_1[r][c] = np.clip(np.matmul(a_n, s_mat1), 0, 1)
            img_mat_2[r][c] = np.clip(np.matmul(a_n, s_mat2), 0, 1)
            img_mat_3[r][c] = np.clip(np.matmul(a_n, s_mat3), 0, 1)

    return img_mat_1, img_mat_2, img_mat_3
    
```

```python
# For each of the two albedos, render three images. 
# One for each of the two light sources, and one for both light-sources combined.

light_1 = light_source[0] / np.linalg.norm(light_source[0])
light_2 = light_source[1] / np.linalg.norm(light_source[1])
light_3 = (light_source[0] + light_source[1]) / np.linalg.norm(light_source[0] + light_source[1])
lights = np.vstack((light_1, light_2, light_3))

# calculate image matrices
I_1, I_2, I_3 = lambertian(normals, lights, albedo, 1, mask)
uI_1, uI_2, uI_3 = lambertian(normals, lights, uniform_albedo, 1, mask)

# plot the face in 2D
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
# display images for albedo
axs[0, 0].imshow(I_1, cmap='gray')
axs[0, 0].set_title('Albedo Source 1')
axs[0, 1].imshow(I_2, cmap='gray')
axs[0, 1].set_title('Light Source 2')
axs[0, 2].imshow(I_3, cmap='gray')
axs[0, 2].set_title('Combined Light')

# display images for uniform albedo
axs[1, 0].imshow(uI_1, cmap='gray')
axs[1, 0].set_title('Uniform Albedo Source 1')
axs[1, 1].imshow(uI_2, cmap='gray')
axs[1, 1].set_title('Light Source 2')
axs[1, 2].imshow(uI_3, cmap='gray')
axs[1, 2].set_title('Combined Light')

plt.tight_layout()
plt.show()
```
